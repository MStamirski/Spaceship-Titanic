# Observations
* Initial accuracy (before model tuning) varied between 75,7% and 78,9%. Generally speaking one hot encoding method (OHE) for categories encoding gave the best results. The least effective models were RidgeClassifier and LinearSVC. The most effective - LightGBM with target encoding (TE) and CatBoost and LogisticRegression implemented in PyTorch (both with OHE). 
* Accuracy after tuning was also the highest when OHE encoding was used. However model CatBoost was an interesting exception - in this single case target encoding (TE) was more effective.
* The best accuracy (79,1%) resulted from LightGBM with OHE and from CatBoost with TE. LighGBM with OHE used the most features compared to the other models: 40 out of 90. CatBoost gave the same value of accuracy using 6 transformed features out of 14.
